```python
import warnings
from Algorithms.SimulatedAnnelingAlgo import SimulatedAnneling
warnings.simplefilter('ignore')
```

```python
def Quality(X):
    return sum(X**4)
```

```python
model = SimulatedAnneling(
        Quality= Quality,
        epochs= 100,
        dimensions= 10,
        bounds= [-100,100],
        stepsize = 0.1,
        precision = 6
    )
model.train()
model.plot_converge()
```
