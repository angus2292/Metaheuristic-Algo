```python
from Algorithms.SOS import SOS
```

```python
def Quality(X):
    return sum(X**4)
```

```python
model = SOS(
    Quality= Quality,
    pop_size= 100,
    dimension= 10,
    epochs= 100,
    bounds= [-100,100],
    precision= 6
)
model.proceed()
model.plotting()
```
