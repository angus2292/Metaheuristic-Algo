```python
import numpy as np
def Quality(X):
    s1 = 0
    s2 = 1
    s1 += sum((X-100)**2)
    for i in range(len(X)):
        s2 = s2 * np.cos((X[i]-100)/(i+1)**0.5)
    return (1/4000)*s1-s2+1
```

```python
model = SOS(
    Quality=Quality,
    pop_size=30,
    dimension=10,
    epochs= 1000,
    bounds= [-600,600],
    precision=6
)
model.proceed()
model.plotting()
```
